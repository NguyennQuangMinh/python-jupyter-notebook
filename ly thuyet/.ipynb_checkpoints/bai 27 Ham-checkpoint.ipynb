```python
# kiem tra so nhap tu ban phim co chia het cho 3 va 5 hay khong, in ra yes neu dung, in ra NO neu sai

def check(n):
    if n%3==0 and n%5==0:
        return True
    else:
        return False

# code de chay chuong trinh
if __name__ == '__main__':
    #code
    a = int(input())
    if(check(a)):
        print('Yes')
    else:
        print('No')
```
