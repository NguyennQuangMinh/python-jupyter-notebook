```python
from math import *

# tìm kiếm tuyến tính: linear search O(N)
# tìm kiếm nhị phân: Binary search O(logN)

# tư tương tìm kiếm nhị phân trong đoạn [l ,r] => tính mid = (l + r) / 2
# 1. a[m]  == x => True
# 2. a[m] < x => tìm ở bên phải
# 3. a[m] > z => tìn ở bên trái

def binary_search(a, l ,r ,x):
    while l < r:
        m = (l+r)//2
        if a[m] == x:
            return True
        elif a[m] < x:
            l = m + 1
        else:
            r = m - 1
    return False

if __name__ == "__main__":
    a = [0 ]
```
