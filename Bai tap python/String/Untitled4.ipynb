```python
# in ra ky tu xuat hien nhieu nhat va it nhat, neu cung bang nhau thi in ra ky tu co so thu tu tu dien lon hon
if __name__ == "__main__":
    s = input()
    d = dict({})
    for i in s:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
    min_fre, max_fre = 10**18, -10**18
    res, res1 = ' ', ' '
    for kytu in sorted(d):
        if d[kytu] >= max_fre:
            max_fre = d[kytu]
            res = kytu
        elif d[kytu] <= min_fre:
            min_fre = d[kytu]
            res1 = kytu
    print(res, max_fre)
    print(res1, min_fre)
```
