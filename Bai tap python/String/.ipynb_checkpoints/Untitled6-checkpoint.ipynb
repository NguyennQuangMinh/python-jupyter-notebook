```python

if __name__ == "__main__":
    s1 = input()
    s2 = input()
    set1 = set(s1) # lấy ra những ký tự khác nhau trong chuỗi string S1
    set2 = set(s2) # lấy ra những ký tự khác nhau trong chuỗi string S2
    res1 = set1.difference(set2) # lấy ra những ký tự chỉ xuất hiện trong S1 mà ko xuất hiện trong S2
    res2 = set2.difference(set1) # lấy ra những ký tự chỉ xuất hiện trong S2 mà ko xuất hiện trong S1
    print(''.join(sorted(res1))) # in ra màn hình theo thứ tự từ điển
    print(''.join(sorted(res2))) # in ra màn hình theo thứ tự từ điển
```
