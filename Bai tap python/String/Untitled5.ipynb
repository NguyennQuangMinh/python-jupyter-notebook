```python

if __name__ == "__main__":
    s1 = input()
    s2 = input()
    set1 = set(s1) # lấy ra những ky tu da xuat hien trong string s1
    set2 = set(s2) # lấy ra những ky tu da xuat hien trong string s2
    giao = set1.intersection(set2) # lấy giao cua s1 và s2
    hop = set1.union(set2) # lấy hợp của s1 và s2
    print(''.join(sorted(giao))) # ép thành chuỗi sắp xếp theo thứ tự từ điển
    print(''.join(sorted(hop))) # ép thành chuỗi sắp xếp theo thứ tự từ điển
```
