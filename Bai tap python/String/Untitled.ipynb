```python
# đếm tần suất xuất hiện của chuỗi ký tự
from collections import Counter
if __name__ == "__main__":
    s = '28tech python 28tech aaaba'
    # chuyển sang mã ascii sau đó sử dụng mảng đánh dấu
    cnt = [0] * 256
    for i in s:
        cnt[ord(i)] += 1
    for i in s:
        if cnt[ord(i)] != 0:
            print(i, cnt[ord(i)]);
            cnt[ord(i)] = 0
    print()
    # cách 2: sử dụng dictionary
    d = dict({})
    for x in s:
        if x in d:
            d[x] += 1
        else:
            d[x] = 1
    for x,y in d.items():
        print(x, y)
    #cách 3: counter
    q = dict(Counter(s)).items()
    print(q)
```
