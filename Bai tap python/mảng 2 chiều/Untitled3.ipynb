```python
from math import *

def tn(n):
    res = n
    reg = 0
    while(n!=0):
        reg = reg*10 + n%10
        n //= 10
    if reg == res:
        return True
    return False
if __name__ == "__main__":
    n = int(input())
    a = []
    for _ in range(n):
        b = list(map(int, input().split()))
        a.append(b)
    dem = 0
    for i in range(n):
        for j in range(i+1):
            if tn(a[i][j]):
                dem += 1
    print(dem)
```
