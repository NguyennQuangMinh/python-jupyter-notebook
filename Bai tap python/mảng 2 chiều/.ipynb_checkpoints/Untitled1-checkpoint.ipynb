```python
from math import *

if __name__ == "__main__":
    n, m = map(int, input().split())
    a = []
    for i in range(n):
        b = map(int, input().split())
        a.append(b)
    # tổng từng hàng
    for i in range(n):
        print(sum(a[i]), end = ' ')
    print()
    # tổng từng cột
    for j in range(m):
        tong = 0
        for i in range(n):
            tong += a[i][j]
        print(tong, end = ' ') 
```
