```python
# Alice : ban dau co a xu
# Barbara : ban dau co b xu
# Cerene : ban dau co c xu
# n = A+B+C
# a+A = b+B = c+C
a, b, c, n= map(int, input().split())
if (n+a+b+c)%3==0:
    temp = (n+a+b+c)//3
    if temp>=a and temp >=b and temp>=c:
        print('Yes')
    else:
        print('No')
else:
    print('NO')
```
