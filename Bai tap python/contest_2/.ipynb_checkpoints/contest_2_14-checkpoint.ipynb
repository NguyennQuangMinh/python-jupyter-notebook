```python
a, b, c ,d = map(float, input().split())
if a>=0 and a<=10 and b>=0 and b<=10 and c>=0 and c<=10 and d>=0 and d<=10:
    tinh_diem = (a+b+c*2+d*3)/7
    if tinh_diem>=8:
        print('GIOI')
    elif tinh_diem>=6.5:
        print('KHA')
    elif tinh_diem>=5:
        print('TRUNG BINH')
    else:
        print('Yeu')
else:
    print('INVALID')
```
