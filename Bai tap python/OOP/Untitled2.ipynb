```python

class sinhvien:
    def __init__(self, ma, ten, lop, ns, gpa):
        self.ma = ma
        self.ten = ten
        self.lop = lop
        self.ns = ns
        self.gpa = gpa
    def chuanhoa(self):
        if self.ns[1] == '/':
            self.ns = '0' + self.ns
        if self.ns[4] == '/':
            self.ns = self.ns[0:3] + '0' + self.ns[3:]
    def __str__(self):
        return f'{self.ma} {self.ten} {self.lop} {self.ns} {self.gpa:.2f}'

if __name__ == "__main__":
    ten = input()
    lop = input()
    ngaysinh = input()
    gpa = float(input())
    s = sinhvien('0001', ten, lop, ngaysinh, gpa)
    s.chuanhoa()
    print(s)
```
