```python
from math import *

class SINHVIEN:
    def __init__(self, hoten, ngaysinh, m1, m2, m3 ):
        self.hoten = hoten
        self.ngaysinh = ngaysinh
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
    def __str__(self):
        tongdiem = m1+m2+m3
        return f'{self.hoten} {self.ngaysinh} {tongdiem:.1f}'

if __name__ == "__main__":
    ten = input()
    ngaysinh = input()
    m1, m2, m3 = map(float, input().split())
    s = SINHVIEN(ten, ngaysinh, m1, m2, m3)
    print(s)
```
