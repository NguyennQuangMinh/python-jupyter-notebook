```python

class mathang:
    def __init__(self, ma, ten, donvi, mua, ban):
        self.ma = str(ma)
        while len(self.ma) < 4:
            self.ma = '0' + self.ma
        self.ma = 'MH' + self.ma
        self.ten = ten
        self.donvi = donvi
        self.mua = mua
        self.ban = ban
    def __str__(self):
        return f'{self.ma} {self.ten} {self.donvi} {self.mua} {self.ban} {self.ban - self.mua}'
    def get_loi_nhuan(self):
        return self.ban - self.mua
    def get_ma(self):
        return self.ma

if __name__ == "__main__":
    n = int(input())
    a = []
    for i in range(n):
        m = mathang(i+1, input(), input(), int(input()), int(input()))
        a.append()
    a.sort(key = lambda x : -x.get_loi_nhuan())
    for x in a:
        print(x)
```
