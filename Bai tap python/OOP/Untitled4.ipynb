```python

class thisinh:
    def __init__(self, ma, ten, toan, ly , hoa):
        self.ma = ma
        self.ten = ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa
    def khuvuc(self):
        return int(self.ma[2])
    def infor(self):
        print(self.ma + ' ' + self.ten + ' ' + str(self.khuvuc()), end = ' ' )
        tongdiem = self.toan + self.ly + self.hoa
        kv = self.khuvuc()
        if kv == 1:
            tongdiem += 0.25
        elif kv == 2:
            tongdiem += 1
        elif kv == 3:
            tongdiem += 2.5
        if tongdiem == int(tongdiem):
            print(int(tongdiem), end = ' ')
        else:
            print('%.1f' %tongdiem, end = ' ')
        if tongdiem >= 24:
            print('Do')
        else:
            print('Truot')
    
if __name__ == "__main__":
    n = thisinh(input(), input(), float(input()), float(input()), float(input()))
    n.infor()
```
