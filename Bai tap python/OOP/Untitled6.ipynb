```python

class sinhvien:
    def __init__(self, ma, ten, lop, ngaysinh, gpa):
        self.ma = format(ma, '03d')
        self.ma = 'SV' + self.ma
        self.ten = ten
        self.lop = lop
        self.ngaysinh = ngaysinh
        self.gpa = gpa
    def chuanhoa(self):
        if self.ngaysinh[1] == '/':
            self.ngaysinh = '0' + self.ngaysinh
        if self.ngaysinh[4] == '/':
            self.ngaysinh = self.ngaysinh[0:3] + '0' + self.ngaysinh[3:]
        tmp = self.ten.split()
        self.ten = ' '.join(tmp).title()
    def __str__(self):
        return f'{self.ma} {self.ten} {self.lop} {self.ngaysinh} {self.gpa:.1f}'
if __name__ == "__main__":
    n = int(input())
    for i in range(n):
        s = sinhvien(i+1, input(), input(), input(), float(input()))
        s.chuanhoa()
        print(s)
```
