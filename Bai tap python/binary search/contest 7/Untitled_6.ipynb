```python

def binary_search(a, l, r, x):
    while l <= r:
        m = (l+r)//2
        if a[m] == x:
            return True
        elif a[m] < x:
            l = m+1
        else:
            r = m-1
    return False

if __name__ == "__main__":
    n = int(input())
    a = list(map(int, input().split()))
    a.sort()
    q = int(input())
    for _ in range(q):
        m = int(input())
        if binary_search(a, 0, n, m):
            print("YES")
        else:
            print("NO")
        
```
