```python
if __name__ == "__main__":
    n = int(input())
    a = list(map(int, input().split()))
    ans = 1 #bien luu tru tan suat lon nhat
    cnt = 1 #bien dem tan suat
    res = a[0] # bien luu gia tri co tan suat xuat hien nhieu nhat
    a.sort()
    a.append(10**18)
    for i in range(1, n+1):
        if a[i] == a[i-1]:
            cnt += 1
        else:
            if cnt>ans:
                ans = cnt
                res = a[i-1]
            cnt = 1
    print(res, ans)
```
