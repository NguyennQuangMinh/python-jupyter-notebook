```python
from math import *

# tìm vị trí xuất hiện đầu tiên của phần tử có giá trị X trong mảng đã sắp xếp tăng dần
def first_pos(a, l, r, x):
    res = -1
    while l<=r :
        m = (l+r)//2
        if a[m] == x:
            res = m
            r = m-1
        elif a[m] < x:
            l = m+1
        else:
            r = m - 1
    return res

# tìm vị trí cuối cùng của phần tử có giá trị X trong mảng đã sắp xếp tăng dần
def last_pos(a, l, r, x):
    res = -1
    while l <= r:
        m = (l+r)//2
        if a[m] == x:
            res = m
            l=m+1
        elif a[m] < x:
            l = m+1
        else:
            r=m-1
    return res

# tìm vị trí đầu tiên có giá trị >=X trong mảng đã được sắp xếp tăng dần
def first_pos1(a,l,r,x):
    res = -1
    while l<=r:
        m = (l+r)//2
        if a[m] >= x:
            res = m
            r = m -1
        else:
            l = m + 1
    return res

if __name__ == "__main__":
    a = [1, 2, 2 , 3, 3, 4, 4, 5, 6]
    print(first_pos(a, 0, len(a) - 1, 4))
    print(last_pos(a, 0, len(a) - 1, 4))
    print(first_pos1(a, 0, len(a) - 1, 4))
```
