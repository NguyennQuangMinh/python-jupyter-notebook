```python
# tinh tong chu so chan, tinh tong chu so le cua so n sử dụng đệ quy

def chan(n):
    if n<10:
        if n % 2 == 0:
            return n
        else:
            return 0
    else:
        if n % 2 == 0:
            return n % 10 + chan(n//10)
        else:
            return chan(n//10)

def le(n):
    if n<10:
        if n % 2 != 0:
            return n
        else:
            return 0
    else:
        if n % 2 != 0:
            return n%10 + le(n//10)
        else:
            return le(n//10)
if __name__ == '__main__':
    n = int(input())
    print(chan(n))
    print(le(n))
```
