```python
def S(n):
    if n==0:
        return 0
    else:
        if n%2 == 0:
            return n + S(n-1)
        else:
            return -n + S(n-1)
# tính giai thừa bằng đệ quy
def S_giaithua(n):
    if n == 0:
        return 1
    else:
        return n * S_giaithua(n-1)
if __name__ == '__main__':
    a = int(input())
    print(S_giaithua(a))
    
```
