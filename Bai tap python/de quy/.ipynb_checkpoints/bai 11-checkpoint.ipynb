```python
# in ra số lớn nhất và số nhỏ nhất của số n
from math import *
def S1(n):
    if n < 10:
        return n
    else:
        return max(n%10, S1(n//10))

def S2(n):
    if n < 10:
        return n
    else:
        return min(n%10, S2(n//10))

if __name__ == '__main__':
    n = int(input())
    print(S1(n))
    print(S2(n))
```
