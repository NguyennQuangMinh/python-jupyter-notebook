```python
# sử dụng đệ quy in ra từ phải qua trái và từ trái qua phải của số n
def S1(n):
    if n < 10:
        print(n, end = ' ')
    else:
        print(n%10, end = ' ')
        S1(n//10)

def S2(n):
    if n < 10:
        print(n, end = ' ')
    else:
        S2(n//10)
        print(n%10, end = ' ')

if __name__ == '__main__':
    n = int(input())
    S1(n)
    print()
    S2(n)
```
