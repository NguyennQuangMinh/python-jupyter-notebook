```python
# kiểm tra xem tất cả các chữ số trong số n có phải chữ số chẵn hay không

def check(n):
    if n < 10:
        if n%2 == 0:
            return True
        else:
            return False
    else:
        if n%2 !=0: return False
        return(check(n//10))
if __name__ == '__main__':
    n = int(input())
    print(check(n))    
```
