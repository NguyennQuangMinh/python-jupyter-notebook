```python
from math import *
# UCLN: sử dụng thuật toán gcd()
#BCNN = 2 số nhân với nhau rồi chia cho UCLN
def gcd(a, b):
    while b != 0 :
        a, b = b, a%b
    return a

def lcm(a,b):
    return a*b / gcd(a,b)
if __name__ == '__main__':
    print(gcd(18,15))
    print(lcm(18, 15))

""" # tim UCLN va BCNN cua nhieu so thi goi lai ham gcd() va lcm() với số lần tương ứng
if __name__ == '__main__':
    x,y,z = 30, 40 , 10
    print(gcd(gcd(x, y), z))
    print(lcm(lcm(x, y), z)) """
```
