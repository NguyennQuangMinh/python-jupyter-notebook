```python
from math import *
# số thuận nghịch, số đối xứng, 

def thuan_nghich(n):
    res = 0
    temp = n
    while n != 0:
        res = res*10 + n%10
        n //= 10
    if res == temp:
        return True
    else:
        return False
if __name__ == '__main__':
    a = int(input())
    print(thuan_nghich(a))
    
```
