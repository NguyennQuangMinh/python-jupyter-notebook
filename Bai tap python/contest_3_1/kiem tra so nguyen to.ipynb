```python
from math import *
# kiem tra số nguyên tố
# liet ke cac so nguyen to tu 1 cho toi n
# số nguyên tố là số chia hết cho 1 và chính nó, 1 ko phải là số nguyên tố
# khi code kiểm tra tính chất đúng sai => Nên cố gắng chứng minh nó sai => Kết luận bằng câu lệnh return
def prime(n):
    if n<2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n% i == 0:
            return False
    return True
        
if __name__ == '__main__':
    n = int(input())
    #liet ke cac so nguyen to tu 1 toi n
    for i in range(1, n+1):
        if prime(i):
            print(i, end=' ')
```
