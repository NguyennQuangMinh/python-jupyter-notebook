```python
from math import *
def perfect_number(n):
    tong = 0
    temp = n
    for i in range(1, isqrt(n) + 1):
        if n%i == 0:
            tong += i
            if n != n//i:
                tong += n//i
    if tong == temp:
        return True
    else:
        return False
if __name__ == '__main__':
    a = int(input())
    print(perfect_number(a))
    
```
