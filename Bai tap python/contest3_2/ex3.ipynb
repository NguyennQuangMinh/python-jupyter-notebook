```python
from math import *
# tính tổng ước của n
def tong_uoc(n):
    tong = 0
    for i in range(1, isqrt(n) + 1):
        if n%i==0:
            tong +=i
            if i != n//i:
                tong += n//i
    return tong

# kiểm tra số chính phương
def chinh_phuong(n):
    can = isqrt(n)
    if can*can == n:
        return True
    else:
        return False
if __name__ == '__main__':
    cnt  = 0
    a, b = map(int, input().split())
    #in ra số chính phương trong đoạn từ a đến b
    for i in range(isqrt(a), isqrt(b) + 1):
        tmp = i*i
        if tmp >= a and tmp <=b:
            cnt += 1
    print(cnt)
```
