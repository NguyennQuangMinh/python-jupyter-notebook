```python
from math import *
# Hàm 1
def nguyen_to(n):
    if n==0 or n==1:
        return 0
    else:
        for i in range(2, isqrt(n) + 1):
            if n % i == 0:
                return 0
        return 1

#Hàm 2
def tong_chu_so(n):
    tong = 0
    while n!= 0:
        tong += n%10
        n//=10
    return tong

#Ham 3
def tong_chu_so_chan(n):
    tong = 0
    while n!=0:
        if (n%10) % 2 == 0:
            tong += n%10
        n//=10
    return tong

#Ham 4
def tong_chu_so_la_so_nguyen(n):
    tong = 0
    while n!=0:
        if nguyen_to(n%10):
            tong += n%10
        n//=10
    return tong

#Ham 5
def so_lat_nguoc(n):
    lat_nguoc = 0
    while n!= 0:
        lat_nguoc = lat_nguoc*10 + n%10
        n//=10
    return lat_nguoc

# Ham 6
def dem_uoc(n):
    if n < 2:
        return 0
    dem = 0
    for i in range(2, isqrt(n) + 1):
        if n%i == 0:
            dem += 1
            while n%i == 0:
                n//=i
    if n > 1:
        dem += 1
    return dem

# Hàm 7
def uoc_nt_max(n):
    if n < 2:
        return 0
    else:
        res = -1
        for i in range(2, isqrt(n) + 1):
            if n%i ==0:
                res = i
                while n%i ==0:
                    n//=i
    if n>1:
        res = n
    return res

# Hàm 8
def co_so_6(n):
    while n!=0:
        tmp = n%10
        if tmp == 6:
            return 1
        n//=10
    return 0
# Hàm 9
def Ham9(n):
    tong = 0
    while n!=0:
        tong += n%10
        n//=10
    if tong % 8 == 0 :
        return 1
    else:
        return 0

# Hàm 10
def Ham10(n):
    tong = 0
    while n!=0:
        tong += factorial(n%10)
        n//=10
    return tong

# Ham 11
def Ham11(n):
    tmp = n%10
    while n!=0:
        if n%10 != tmp:
            return 0
        n //= 10
    return 1

def Ham12(n):
    tmp = n%10
    while n!=0:
        if n%10 > tmp:
            return 0
        n//=10
    return 1

def Ham13(n):
    mu = 0
    tong = 0
    tmp = n
    while tmp!=0:
        mu += 1
        tmp//=10
    while n!=0:
        tong += (n%10) ** mu
        n//=10
    return tong
if __name__ == "__main__":
    n = 30
    print(uoc_nt_max(n))
```
