```python
from math import *
# số thuần nguyên tố
# kiểm tra số đó có phải là số nguyên tố hay không
def check(n):
    if n<2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n%i ==0:
            return False
    return True
# tổng các chữ số có phải là số nguyên tố hay không và kiếm tra xem tất cả các chữ số có phải số nguyên tố hay không
def sum_check(n):
    tong = 0
    while n!=0:
        r = n%10
        if r!=2 and r!=3 and r!=5 and r!=7:
            return False
        tong += r
        n//=10
    return check(tong)
if __name__ == '__main__':
    cnt = 0
    a, b = map(int, input().split())
    for i in range(a, b+1):
        if sum_check(i) and check(i):
            cnt += 1
    print(cnt)
    
```
