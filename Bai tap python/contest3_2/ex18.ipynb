```python
from math import *
#kiểm tra 1 số có phải là số fibonacci hay không
def check_fibo(n):
    if n==0 or n==1:
        return True
    else:
        fn1=1
        fn2=0
        for i in range(2, 100):
            fn = fn1+fn2
            if fn == n:
                return True
            fn2, fn1 = fn1, fn
        return False
if __name__ == '__main__':
    a = int(input())
    if check_fibo(a):
        print('Yes')
    else:
        print('No')
```
