```python
from math import *
# kiểm tra xem có phải là số lộc phát hay không (0,6,8)
def loc_phat(n):
    while n!=0:
        r = n%10
        if r!=0 and r!=6 and r!=8:
            return False
        n//=10
    return True
if __name__ == '__main__':
    a = int(input())
    if loc_phat(a):
        print(1)
    else:
        print(0)
```
