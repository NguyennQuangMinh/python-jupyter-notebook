```python
from math import *
# kiểm tra xem số này có phải là số đẹp hay không
# kiểm tra xem có phải số thuận nghịch hay không
def thuan_nghich(n):
    tong = 0
    tmp = n
    while n!=0:
        tong = tong*10 + n%10
        n//=10
    if tong == tmp:
        return True
    else:
        return False
# tách thừa số nguyên tố
def check(n):
    cnt = 0
    for i in range(2, isqrt(n) + 1):
        if n%i ==0:
            cnt += 1
            while n%i == 0:
                n//=i
    if n > 1:
        cnt += 1
    return cnt>=3
if __name__ == '__main__':
    dem = 0
    a, b = map(int, input().split())
    for i in range(a, b+1):
        if thuan_nghich(i) and check(i):
            print(i, end =' ')
            dem+=1
    if dem == 0:
        print(-1)
```
