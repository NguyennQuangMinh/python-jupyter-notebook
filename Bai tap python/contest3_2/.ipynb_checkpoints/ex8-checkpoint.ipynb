```python
from math import *

def check_1(n):
    donvi = n%10
    while n!=0:
        if n%10 > donvi:
            return False
        n//=10
    return True
def check_2(n):
    if n<2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n%i ==0:
            return False
    return True
if __name__ == '__main__':
    cnt = 0
    a = int(input())
    for i in range( a + 1):
        if check_1(i) and check_2(i):
            cnt += 1
            print(i, end = ' ')
    print("\n",cnt, sep ='')
```
