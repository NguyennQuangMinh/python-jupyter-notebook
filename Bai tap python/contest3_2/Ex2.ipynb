```python
from math import *
# kiểm tra số nguyên tố
def nguyen_to(n):
    if n<2:
        return False
    else:
        for i in range(2, isqrt(n) + 1):
            if n%i == 0:
                return False
        return True

# phân tích ra thừa số nguyên tố rồi kiểm tra có phải là số sphenic hay không
def sphenic(n):
    cnt = 0
    for i in range(2, isqrt(n)):
        if n % i == 0:
            mu = 0
            while n%i == 0:
                mu += 1
                n//=i
            if mu > 1:
                return False
            cnt+=1
    if n > 1:
        cnt += 1
    return cnt == 3

# phân tích thừa số nguyên tố
def pt(n):
    for i in range(2, isqrt(n)):
        if n%i == 0:
            mu = 0
            while n % i == 0:
                mu += 1
                n//=i
            print(i, mu, sep ='^', end ='')
            if n!=1: # chua phan tich xong
                print('*', end='')
    if n>1:
        print(n, 1, sep = '^')

# tim Uoc so nguyen to lon nhat cua n
def solve(n):
    ans = -1 #khoi tao bien luu tru uoc so nguyen to lon nhat bang 1 gia tri sai
    for i in range(2, isqrt(n) + 1):
        if n%i == 0:
            ans = i
            while n%i ==0:
                n//=i
    if n>1:
        ans = n
    return ans

# kiểm tra số Smith
def sum_digit(n):
    tong = 0
    while n!=0:
        tong += n%10
        n//=10
    return tong
def check(n):
    sum1 = sum_digit(n)
    sum2 = 0
    tmp = n
    for i in range(2, isqrt(n)+1):
        if n % i == 0:
            sum2 += sum_digit(i)
            n//=i
    if tmp == n:
        return False
    if n > 1:
        sum2 += sum_digit(n)
    return sum2 == sum1
if __name__ == "__main__":
    n = int(input())
    if check(n):
        print('Yes')
    else:
        print('No')
```
