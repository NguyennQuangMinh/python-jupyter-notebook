```python
from math import *

def nt(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n%i==0:
            return False
    return True

def tong_chu_so(n):
    tong = 0
    while n!=0:
        tong += n%10
        n//=10
    return tong

def fibo(n):
    fn2 = 0
    fn1 = 1
    for i in range(2, 100):
        fn = fn1 + fn2
        if tong_chu_so(n) == fn:
            return True
        fn2, fn1 = fn1, fn
    return False

if __name__ == '__main__':
    a = int(input())
    for i in range(1, a+1):
        if nt(i) and fibo(i):
            print(i, end = ' ')
    
```
