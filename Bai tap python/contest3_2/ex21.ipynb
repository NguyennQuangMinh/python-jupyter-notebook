```python
def check_fibo(n):
    if n == 0 or n==1:
        return True
    fn1 = 1
    fn2 = 0
    for i in range(2, 100):
        fn = fn1 + fn2
        if n == fn:
            return True
        fn2, fn1 = fn1, fn
    return False

if __name__ == '__main__':
    t = int(input())
    for i in range(1, t+1):
        a = int(input())
        if check_fibo(a):
            print('Yes')
        else:
            print('No')
```
