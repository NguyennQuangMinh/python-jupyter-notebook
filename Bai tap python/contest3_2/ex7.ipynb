```python
from math import *

# số thuận nghịc lộc phát
def check(n):
    tmp = n
    rev = 0
    tong = 0
    ok = False
    while n!=0:
        rev = rev*10 + n%10
        tong += n%10
        if n%10 == 6:
            ok = True
        n//=10
    return ok and (rev == tmp)  and (tong%10 == 8)
    
if __name__ == '__main__':
    cnt = 0
    a, b = map(int, input().split())
    for i in range(a, b+1):
        if check(i):
            print(i, end = ' ')
            cnt +=1
    if cnt == 0:
        print(-1)
```
