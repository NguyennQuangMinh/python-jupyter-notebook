```python
from math import *

# sàng số nguyên tố Eratosthenes: O(N * loglog(N))

# bước 1: coi tất cả các số trong mảng đều là số nguyên tố prime[i] = True
# bước 2: Bội của 1 số nguyên tố sẽ không phải là số nguyên tố

prime = [True] * (10**6 + 1)
#hàm sàng số nguyên tố
def sieve():
    prime[0] , prime[1] = False, False
    for i in range(2, isqrt(10**6 + 1) + 1):
        if prime[i]:
            # duyệt bội từ i*i đến cuối, bước nhảy i
            for j in range(i*i, 10**6 +1, i):
                prime[j] = False
        
if __name__ == '__main__':
    sieve()
    for i in range(100):
        if prime[i]:
            print(i, end = " ")
```
