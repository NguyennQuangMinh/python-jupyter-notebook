```python
from math import *

if __name__ == "__main__":
    n = int(input())
    a = list(map(int, input().split()))
    ok = False
    for i in range(n):
        if a[i] == a[0] or a[i] == a[n-1]:
            ok = False
        elif a[i] > a[i-1] and a[i] > a[i+1]:
            ok = True
        else:
            ok = False
        if ok == True:
            print(a[i], end = '\n')
```
