```python
from math import *
# in ra số lượng số chẵn, số lượng số lẻ, tổng chẵn, tổng lẻ trong list
if __name__ == '__main__':
    n = int(input())
    a = list(map(int, input().split()))
    chan, le, tongchan, tongle = 0, 0, 0, 0
    for x in a:
        if x%2 == 0:
            chan += 1
            tongchan += x
        else:
            le += 1
            tongle += x
    print(chan, le, tongchan, tongle, sep = "\n")
    
```
