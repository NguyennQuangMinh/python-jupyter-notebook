```python
# in ra phần tử ở vị trí chẵn và là số chẵn

if __name__ == '__main__':
    n = int(input())
    a = list(map(int, input().split()))
    ok = False
    for i in range(n):
        if i%2 == 0 and a[i] % 2 == 0:
            print(a[i], end = " ")
            ok = True
    if ok == False:
        print("NONE")
```
