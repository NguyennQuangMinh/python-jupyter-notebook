```python

if __name__ == '__main__':
    n = int(input())
    a = list(map(int, input().split()))
    cnt = [0] * 10000001
    for x in a:
        cnt[x] += 1
    cnt2 = cnt.copy()
    for x in a:
        if cnt[x] != 0:
            print(x, cnt[x])
            cnt[x] = 0
    # in theo thứ tự xuất hiện kèm tần suất xuất hiện của số đó
    l ,r = min(a), max(a)
    print()
    for i in range(l, r+1):
        if cnt2[i] != 0:
            print(i, cnt2[i])
    # tìm số xuất hiện nhiều nhất, nếu có cùng số lần xuất hiện nhiều nhất bằng nhau thì in ra số lớn hơn
    max_fre = 0
    max_val = 0
    for i in range(l, r+1):
        if max_fre <= cnt2[i]:
            max_fre = cnt2[i]
            max_val = i
    #tìm số lần xuất hiện ít nhất, nếu có cùng số lần xuất hiện thì in ra số nhỏ hơn
    min_fre = 10**18
    min_val = 10**18
    for i in range(l, r+1):
        if cnt2[i] !=0 and min_fre > cnt2[i]:
            min_fre = cnt2[i]
            min_val = i
    print()
    print(max_val, max_fre)
    print()
    print(min_val, min_fre)
```
