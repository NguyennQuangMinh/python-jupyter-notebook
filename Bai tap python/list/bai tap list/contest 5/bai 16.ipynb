```python
from math import *

#kiểm tra số nguyên tố
def nt(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

# kiểm tra số thuận nghịch
def tn(n):
    if n < 10:
        return False
    tmp = n
    res = 0
    while n != 0:
        res = res*10 + n%10
        n//=10
    if res == n:
        return True
    else:
        return False

# kiểm tra số chính phương
def cp(n):
    tmp = n
    res = int(isqrt(n))
    if res*res == tmp:
        return True
    else:
        return False

# tính tổng các chữ số, kiểm tra xem tổng các chữ số có phải là số nguyên tố hay không
def tong_nt(n):
    res = 0
    if n < 10:
        res = n
    while n!=0:
        res += n%10
        n//=10
    if nt(res):
        return True
    else:
        return False
    
if __name__ == "__main__":
    n = int(input())
    a = list(map(int, input().split()))
    d1, d2, d3, d4 = 0, 0, 0, 0
    for x in a:
        if nt(x):
            d1 += 1
        if tn(x):
            d2 += 1
        if cp(x):
            d3 += 1
        if tong_nt(x):
            d4 += 1
    print(d1, d2, d3, d4, sep = '\n')
    
```
