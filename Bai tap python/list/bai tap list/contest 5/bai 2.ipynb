```python
from math import *
# in ra số lượng số nguyên tố trong mảng, tổng các số nguyên tố trong mảng, trung bình cộng của các số nguyên tố trong mảng
def nt(n):
    if n < 2:
        return False
    for i in range(0, isqrt(n) + 1):
        if n%2 == 0:
            return False
    return True
if __name__ == '__main__':
    n = int(input())
    a = list(map(int, input().split()))
    dem, tong, tbc = 0, 0, 0
    for x in a:
        if nt(x):
            dem += 1
            tong += x
    tbc = tong / dem
    print(dem, tong, sep = " ")
    print('%.3f' %tbc)
```
