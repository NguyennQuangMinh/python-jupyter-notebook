```python
from math import *

if __name__ == '__main__':
    n = int(input())
    a = list(map(int, input().split()))
    dem_chan = 0
    dem_le = 0
    for i in range(n):
        if a[i] % 2 == 0:
            dem_chan += 1
        else:
            dem_le +=1
    if dem_chan > dem_le:
        print('chan')
    elif dem_le > dem_chan:
        print('le')
    else:
        print('chan le')
            
```
