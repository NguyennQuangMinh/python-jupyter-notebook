```python
from math import *

def nt(n):
    if n<2:
        return False
    for i in range(0, isqrt(n) + 1):
        if n%2 == 0:
            return False
    return True

if __name__ == '__main__':
    i = int(input())
    a = list(map(int, input().split()))
    tong_truoc = 0
    tong_sau = 0
    for x in a:
        if x < i:
            tong_truoc += a[x]
        if x > i:
            tong_sau += a[x]
    if nt(tong_truoc):
        print("")
```
