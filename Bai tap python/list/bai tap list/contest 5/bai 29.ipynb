```python
from math import *
# in ra số có số lần xuất hiện nhiều nhất trong mảng, nếu có nhiều số cùng tần suất thì in ra số nhỏ nhất
if __name__ == '__main__':
    n = int(input())
    a = list(map(int, input().split()))
    max_pre, min_val = 0, 10**18
    cnt = [0] * 1000001
    for x in a:
        cnt[x] += 1
    l, r = min(a), max(a)
    for i in range(l, r+1):
        if max_pre < cnt[i]:
            max_pre = cnt[i]
            if i < min_val:
                min_val = i
    print(min_val, max_pre)
```
