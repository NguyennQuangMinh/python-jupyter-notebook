```python
from math import *
# dạng 4 mảng đánh dấu:
# giá trị khác nhau trong mảng: tìm số lượng phần từ khác nhau, liệt kê
# Những bài toán liên quan đến tần suất

# cho mảng A[] gồm n phần tử, đếm xem mỗi phần tử trong mảng xuất hiện bn lần
# chỉ áp dụng cho số >=0 và <=10^7

if __name__ == '__main__':
    a = [4, 2, 1, 2, 3, 1, 4, 2]
    cnt = [0] * (10**7+1)
    for x in a:
        cnt[x] = 1
    for x in a:
        if cnt[x] != 0:
            print(x, cnt[x])
            cnt[x] = 0
```
