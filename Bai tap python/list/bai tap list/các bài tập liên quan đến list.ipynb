```python
from math import *
# dang 1 kiem tra cac phan tu trong list thoa man tính chất nào đó
# dạng 2: tìm Max, tìm Min
# dạng 3: kiểm tra các cặp phần tử trong list thỏa mãn điều kiện nào đó
def nt(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n)+1):
        if n % i == 0:
            return False
    return True

if __name__ == '__main__':
    n = int(input())
    a = list(map(int, input().split()))
    for i in range(n):
        for j in range(i+1, n):
            print(a[i], a[j])
        
```
